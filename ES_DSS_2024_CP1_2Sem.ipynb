```python
++import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
base = pd.read_csv('/content/sample_data/netflix_titles.csv')
base2 = pd.read_csv('/content/sample_data/rotten_tomatoes_critic_reviews.csv')
base3 = pd.read_csv('/content/sample_data/imdb_top_1000.csv')
df = base.dropna()
df
```

```python
df_netflix_terror = df.loc[(df['listed_in'].str.contains('Horror')) & (df['type'] == 'Movie')]
```

```python
# Contagem total
total_terror_movies = df_netflix_terror.shape[0]
print(f'Total de filmes de terror: {total_terror_movies}')

# Distribuição por ano
df_netflix_terror['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(12, 6))
plt.xlabel('Ano')
plt.ylabel('Número de Filmes')
plt.title('Número de Filmes de Terror por Ano')
plt.show()
```

```python


# Filtrar os filmes de terror
df_terror_base3 = base3[base3['Genre'].str.contains('Horror', case=False, na=False)]

# Agrupar por ano e calcular a média das avaliações
avg_rating_by_year = df_terror_base3.groupby('Released_Year')['IMDB_Rating'].mean().reset_index()

# Gráfico de linha mostrando a evolução das avaliações médias por ano
plt.figure(figsize=(12, 6))
plt.plot(avg_rating_by_year['Released_Year'], avg_rating_by_year['IMDB_Rating'], marker='o', linestyle='-')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Avaliação Média IMDb')
plt.title('Avaliação Média dos Filmes de Terror por Ano de Lançamento')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtrar filmes de terror
df_netflix_terror = df.loc[(df['listed_in'].str.contains('Horror')) & (df['type'] == 'Movie')]

# Contar o número de filmes de terror por país
film_count_by_country = df_netflix_terror['country'].value_counts()

# Selecionar os 5 países com o maior número de filmes de terror
top_5_countries = film_count_by_country.head(5)

# Gráfico de barras
plt.figure(figsize=(12, 6))
top_5_countries.plot(kind='bar', color='skyblue')
plt.xlabel('País')
plt.ylabel('Número de Filmes de Terror')
plt.title('Top 5 Países com Mais Filmes de Terror na Netflix')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtrar os filmes de terror
df_terror_base3 = base3[base3['Genre'].str.contains('Horror', case=False, na=False)]

# Filtrar outros gêneros (exemplo: excluindo terror)
df_other_genres_base3 = base3[~base3['Genre'].str.contains('Horror', case=False, na=False)]

# Teste t de Student
t_stat, p_value = stats.ttest_ind(df_terror_base3['IMDB_Rating'].dropna(), df_other_genres_base3['IMDB_Rating'].dropna())

print(f"Estatística t: {t_stat:.2f}")
print(f"Valor p: {p_value:.4f}")

# Boxplot para comparar as distribuições
plt.figure(figsize=(12, 6))
plt.boxplot([df_terror_base3['IMDB_Rating'].dropna(), df_other_genres_base3['IMDB_Rating'].dropna()],
            labels=['Terror', 'Outros Gêneros'])
plt.ylabel('Avaliação IMDb')
plt.title('Comparação das Avaliações IMDb entre Filmes de Terror e Outros Gêneros')
plt.show()
```

```python
# Adicionar a duração média dos filmes por ano
df_terror_base3['Runtime'] = df_terror_base3['Runtime'].str.extract('(\d+)').astype(float)  # Extrair duração em minutos
duration_by_year = df_terror_base3.groupby('Released_Year')['Runtime'].mean()

# Gráfico de linha mostrando a evolução da duração
plt.figure(figsize=(12, 6))
duration_by_year.plot(kind='line', marker='o')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Duração Média (min)')
plt.title('Evolução da Duração dos Filmes de Terror ao Longo dos Anos')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Assumindo que você tem uma coluna 'Critic_Rating' e 'Audience_Rating' (ou substituto)
df_terror_base3['Rating_Difference'] = df_terror_base3['IMDB_Rating'] - df_terror_base3['Meta_score']

# Teste t de Student
t_stat_diff, p_value_diff = stats.ttest_1samp(df_terror_base3['Rating_Difference'].dropna(), 0)

print(f"Estatística t: {t_stat_diff:.2f}")
print(f"Valor p: {p_value_diff:.4f}")

# Gráfico de dispersão
plt.figure(figsize=(12, 6))
plt.scatter(df_terror_base3['IMDB_Rating'], df_terror_base3['Rating_Difference'], alpha=0.5)
plt.xlabel('Avaliação IMDb')
plt.ylabel('Diferença entre Crítica e Audiência')
plt.title('Diferença entre Crítica e Audiência para Filmes de Terror')
plt.grid(True)
plt.show()
```

```python
# Calcular a média das avaliações por diretor
director_avg_rating = df_terror_base3.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False)

# Contar o número de filmes de terror por diretor
director_count = df_terror_base3['Director'].value_counts()

# Filtrar diretores com mais de um filme para uma análise mais significativa
directors_filtered = director_count[director_count > 1].index
filtered_avg_rating = director_avg_rating[directors_filtered]

# Gráfico de barras mostrando a média das avaliações por diretor
plt.figure(figsize=(12, 6))
filtered_avg_rating.plot(kind='bar')
plt.xlabel('Diretor')
plt.ylabel('Avaliação Média IMDb')
plt.title('Avaliação Média dos Filmes de Terror por Diretor')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assumindo que 'Meta_score' é a avaliação da crítica e 'IMDB_Rating' é a avaliação do público
df_terror_base3['Rating_Discrepancy'] = df_terror_base3['IMDB_Rating'] - df_terror_base3['Meta_score']

# Filtrar filmes com a maior discrepância
top_discrepancy = df_terror_base3[['Series_Title', 'Rating_Discrepancy']].sort_values(by='Rating_Discrepancy', ascending=False).head(10)

# Gráfico de barras mostrando as maiores discrepâncias
plt.figure(figsize=(12, 6))
plt.barh(top_discrepancy['Series_Title'], top_discrepancy['Rating_Discrepancy'])
plt.xlabel('Discrepância entre Avaliação da Crítica e do Público')
plt.ylabel('Filme')
plt.title('Filmes de Terror com Maior Discrepância nas Avaliações')
plt.show()
```

```python
# Filtrar filmes de terror
df_terror_base3 = base3[base3['Genre'].str.contains('Horror', case=False, na=False)]

# Converter a bilheteira para valores numéricos
df_terror_base3['Gross'] = df_terror_base3['Gross'].replace('[\$,]', '', regex=True).astype(float)

# Somar a bilheteira por ano
total_gross_by_year = df_terror_base3.groupby('Released_Year')['Gross'].sum()

# Plotar o gráfico de barras
plt.figure(figsize=(12, 6))
total_gross_by_year.plot(kind='bar')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Bilheteira Total (em milhões)')
plt.title('Bilheteira Total dos Filmes de Terror por Ano')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
